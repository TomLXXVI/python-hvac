# Analysis of Single-Stage Vapor Compression Machine

Analyzing the influence of **outdoor air temperature** entering the condenser and **condenser air mass flow rate** (= outdoor air mass flow rate) on steady-state machine performance.

Fixed operating conditions of the machine:
- entering air state and mass flow rate at evaporator: 24 Â°C, 50 % RH, 1500 kg/hr
- set degree of superheat: 10 K

We will use the analysis-functions from modules `ex07_multi_analysis_2b.py` (for analyzing the influence of outdoor air temperature) and `ex07_multi_analysis_3b.py` (for analyzing the influence of outdoor air mass flow rate):

```python
import ex07_multi_analysis_2b as oat_analysis
import ex07_multi_analysis_3b as oam_analysis
```

```python
# get simulation data with varying compressor speed and varying outdoor air temperature
df_oat = oat_analysis.read_dataframe()

# get simulation data with varying compressor speed and varying condenser air mass flow rate
df_oam = oam_analysis.read_dataframe()
```

## Influence of Outdoor Temperature

The mass flow rate of air through the condenser is fixed at 2233.861 kg/hr.

### Refrigeration Capacity

**Refrigeration capacity as function of outdoor temperature at given compressor speeds**

```python
oat_analysis.plot_Q_evp(df_oat, n_cmp_rng=[3000, 3300, 3600, 4500, 4800, 5100])
```

**Refrigeration capacity as function of compressor speed at given outdoor temperatures**

```python
oat_analysis.plot_Q_evp(df_oat, T_out_rng=[26, 30, 34, 38]) 
```

**Observations**
- Refrigeration capacity decreases somewhat with increasing outdoor temperature at a given compressor speed.
- Refrigeration capacity increases with increasing compressor speed at a given outdoor temperature.

### Compressor Power

**Compressor power as function of outdoor temperature at given compressor speeds**

```python
oat_analysis.plot_W_cmp(df_oat, n_cmp_rng=[3000, 3300, 3600, 3900, 4200, 4500])
```

**Compressor power as function of compressor speed at given outdoor temperatures**

```python
oat_analysis.plot_W_cmp(df_oat, T_out_rng=[26, 30, 34, 38])
```

**Observations**
- Compressor power remains fairly constant with increasing outdoor temperature at a given compressor speed.
- Compressor power increases almost linearly with increasing compressor speed at a given outdoor temperature.

### Heat Rejection Rate

**Heat rejection rate as function of outdoor temperature at given compressor speeds**

```python
oat_analysis.plot_Q_cnd(df_oat, n_cmp_rng=[3000, 3300, 3600, 3900, 4200, 4500])
```

**Heat rejection rate as function of compressor speed at given outdoor temperatures**

```python
oat_analysis.plot_Q_cnd(df_oat, T_out_rng=[26, 30, 34, 38])
```

**Observations**
- Heat rejection rate slightly decreases with increasing outdoor temperature at a given compressor speed.
- Heat rejection rate increases with increasing compressor speed at a given outdoor temperature.

## Influence of Outdoor Air Mass Flow Rate

Outdoor air temperature remains fixed at 30 Â°C.

### Refrigeration Capacity

**Refrigeration capacity as function of outdoor air mass flow rate at given compressor speeds**

```python
oam_analysis.plot_Q_evp(df_oam, n_cmp_rng=[3000, 3300, 3600, 3900, 4200, 4500])
```

**Refrigeration capacity as function of compressor speed at given outdoor air mass flow rates**

```python
oam_analysis.plot_Q_evp(df_oam, cnd_m_dot_air_rng=[2000, 2500, 3000, 3500, 4000])
```

**Observations**
- Refrigeration capacity very slightly increases with increasing outdoor air mass flow rate at a given compressor speed. The increase is more pronounced at higher compressor speeds.
- Refrigeration capacity increases with increasing compressor speed at a given outdoor air mass flow rate. At lower compressor speeds refrigeration capacity is almost independent of outdoor air mass flow rate.

### Compressor Power

**Compressor power as function of outdoor air mass flow rate at given compressor speeds**

```python
oam_analysis.plot_W_cmp(df_oam, n_cmp_rng=[3000, 3300, 3600, 3900, 4200, 4500])
```

**Compressor power as function of compressor speed at given outdoor air mass flow rates**

```python
oam_analysis.plot_W_cmp(df_oam, cnd_m_dot_air_rng=[2000, 2500, 3000, 3500, 4000])
```

**Observations**
- Compressor power is almost independent of outdoor air mass flow rate.
- Compressor power increases linearly with compressor speed.

### Heat Rejection Rate

**Heat rejection rate as function of outdoor air mass flow rate at given compressor speeds**

```python
oam_analysis.plot_Q_cnd(df_oam, n_cmp_rng=[3000, 3300, 3600, 3900, 4200, 4500])
```

**Heat rejection rate as function of compressor speed at given outdoor air mass flow rates**

```python
oam_analysis.plot_Q_cnd(df_oam, cnd_m_dot_air_rng=[2000, 2500, 3000, 3500, 4000])
```

**Observations**
- At lower compressor speeds, heat rejection rate is almost independent of outdoor air mass flow rate. At higher compressor speeds, the heat rejection rate can be increased somewhat by increasing the condenser air mass flow rate.
- Heat rejection rate increases with compressor speed at a given mass flow rate of condenser air. The increase of the heat rejection rate becomes steeper as the condenser air mass flow rate is higher.
